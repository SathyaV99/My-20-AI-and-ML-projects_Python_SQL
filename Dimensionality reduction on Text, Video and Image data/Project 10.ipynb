<font size="5"><p style="color:black"> <b> 1. List down all possible dimensionality reduction techniques that can be implemented using python.  

<font size="4"> Dimensionality Reduction is a process of reducing dimensions or features of the dataset. The dimensions of the dataset can be reduced by using two ways. They are:

* Feature Selection - Feature Selection is the process of selecting dimensions of features of the dataset which contributes mode to the machine learning tasks such as classification, clustering, e.t.c. This can be achieved by using different methods such as correlation analysis, univariate analysis, e.t.c.

* Feature Extraction - Feature Extraction Feature Extraction is a process of finding new features by selecting and/or combining existing features to create reduced feature space, while still accurately and completely describing the data set without loss of information.

1) <b> Principal Component Analysis(PCA) </b> - PCA is a linear unsupervised dimensionality reduction technique that can be described as an eigen decomposition of the covariance matrix of the data. It increases the interpretability by reducing the information loss by creating uncorrelated features with maximized variance.

2) <b>Independent Component Analysis (ICA) </b>  - It is a machine learning technique to separate independent sources from a mixed signal. Unlike principal component analysis which focuses on maximizing the variance of the data points, the independent component analysis focuses on independence, i.e. independent components.

3) <b>Linear Discriminant Analysis (LDA)</b> - It is a predictive modeling algorithm for multi-class classification. It can also be used as a dimensionality reduction technique, providing a projection of a training dataset that best separates the examples by their assigned class.

4)<b>Kernel PCA (KPCA)</b> - It uses a kernel function to project dataset into a higher dimensional feature space, where it is linearly separable. 

5) <b>Isomap</b> - It is an Unsupervised Machine Learning technique aimed at Dimensionality Reduction. It differs from a few other techniques in the same category by using a non-linear approach for dimensionality reduction instead of linear mappings used by algorithms such as PCA. 

6) <b>Locally Linear Embedding (LLE)</b> - Locally Linear Embedding (LLE) is an unsupervised method for non-linear dimensionality reduction that can discover non-linear structures in the data set, and also preserve the distances within local neighborhoods.

7) <b>Low Variance Filter</b> - It is a useful dimensionality reduction algorithm that removes values with low variance using conditional statements and loops. 

8) <b>High Correlation filter</b> - This dimensionality reduction algorithm discards attributes with high correlation. It can also be performed with conditional statements and loops.

9) <b>Variance Inflation Factor (VIF)</b> - The VIF measures the severity of multicollinearity in regression analysis. It is a statistical concept that indicates the increase in the variance of a regression coefficient as a result of collinearity.

10) <b>t- Distributed Stochastic Neighbor Embedding (t-SNE)</b> - It is a nonlinear dimensionality reduction technique well-suited for embedding high-dimensional data for visualization in a low-dimensional space of two or three dimensions.

11) <b>UMAP</b> - It is a dimension reduction technique that can preserve as much of the local, and more of the global data structure as compared to t-SNE, with a shorter runtime. 

12) <b>Random Forest/Decision Tree (Feature Importance)</b> - This is one of the most commonly used techniques which tells us the importance of each feature present in the dataset. We can find the importance of each feature and keep the top most features, resulting in dimensionality reduction

13) <b>Hypothesis testing using P-value</b> - This technique uses hypothesis testing where it compares the p value of target attribute to all features in the dataset and the features which are below the significane level are removed from the dataset.

14) <b>Lasso (least absolute shrinkage and selection operator)</b> - It is a regression analysis method that performs both dimensionality reduction via variable selection and regularization in order to enhance the prediction accuracy and interpretability of the resulting statistical model. </font>

### Yes it is possible to perform dimensionality reduction on Images, Video and Text data. It is as follows:

```python
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.decomposition import PCA
from PIL import Image
import urllib.request as urllib
import math
import os
```

<span style="font-family: Arial; font-weight:bold;font-size:1.6em;color:#0000FF;">2.1 Image data reduction

```python
fd = urllib.urlopen("https://upload.wikimedia.org/wikipedia/commons/thumb/f/f4/Earth-BlueMarble-1997.jpg/1024px-Earth-BlueMarble-1997.jpg")
im = Image.open(fd)
im.save('url.jpg','jpeg')
```

```python
file_size8 = os.stat('url.jpg')
print("Size of file :", file_size8.st_size, "bytes")
```

```python
bytes = file_size8.st_size
kilobytes = 0.001 * bytes
megabyes = 0.000001* bytes
print('KB:',kilobytes)
print('MB:',megabyes)
```

```python
im.size
```

```python
im
```

```python
x, y = im.size
x2, y2 = math.floor(x-800), math.floor(y-800)
im1 = im.resize((x2,y2),Image.ANTIALIAS)
im1.save("image_scaled.jpg",quality=35)
```

```python
im1.size
```

```python
file_size9 = os.stat('image_scaled.jpg')
print("Size of file :", file_size9.st_size, "bytes")
```

```python
bytes = file_size9.st_size
kilobytes = 0.001 * bytes
megabyes = 0.000001* bytes
print('KB:',kilobytes)
print('MB:',megabyes)
```

### Image size reduced from 305.5 kb to 7.804 kb after reduction

<span style="font-family: Arial; font-weight:bold;font-size:1.6em;color:#0000FF;">2.2 Video data reduction

```python
import urllib.request
import cv2
import moviepy.editor as mp
from ipywidgets import Video
```

```python
cap = cv2.VideoCapture('Its Rewind Time.mp4')
```

```python
file_size = os.stat('Its Rewind Time.mp4')
print("Size of file :", file_size.st_size, "bytes")
```

```python
bytes = file_size.st_size
kilobytes = 0.001 * bytes
megabyes = 0.000001* bytes
print('KB:',kilobytes)
print('MB:',megabyes)
```

```python
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('AfterReduction.avi',fourcc, 5, (256,144))
 
while True:
    ret, frame = cap.read()
    if ret == True:
        b = cv2.resize(frame,(256,144),fx=0,fy=0, interpolation = cv2.INTER_CUBIC)
        out.write(b)
    else:
        break
    
cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
file_size1 = os.stat('AfterReduction.avi')
print("Size of file :", file_size1.st_size, "bytes")
bytes = file_size1.st_size
kilobytes = 0.001 * bytes
megabyes = 0.000001* bytes
print('KB:',kilobytes)
print('MB:',megabyes)
```

### Video size reduced from 2.63 mb to 1.927 mb after reduction
