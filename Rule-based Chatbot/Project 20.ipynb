**Checking for Missing Values**

```python
def missing_check(df):
    total = df.isnull().sum().sort_values(ascending=False)   # total number of null values
    percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)  # percentage of values that are null
    missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])  # putting the above two together
    return missing_data # return the dataframe
missing_check(data)
```

## Data Pre-Processing

```python
nltk.download('stopwords') # Downloading stopwords
nltk.download('punkt') # Downloading tokenizer
nltk.download('wordnet') #Downloading wordnet
```

**TOKENISATION**

```python
words=[]
tags = []
doc_words = []
doc_tags = []
```

```python
for intent in data['intents']:
    for pattern in intent['patterns']:
        words_temp = nltk.word_tokenize(pattern)
        words.extend(words_temp)
        doc_words.append(words_temp)
        doc_tags.append(intent['tag'])
        if intent['tag'] not in tags:
            tags.append(intent['tag'])
```

**STEMMING**

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()

words = [stemmer.stem(w.lower()) for w in words if w!="?"]
words = sorted(list(set(words)))
tags = sorted(tags)
```

```python
import pickle

pickle.dump(words,open('words.pkl','wb'))
pickle.dump(tags,open('tags.pkl','wb'))
```

**Displaying the pre processed words and tags**

```python
words = sorted(list(set(words)))
tags = sorted(list(set(tags)))
```

```python
print (len(doc_words),"document words:\n")
print (len(tags), "tags:\t", tags,'\n')
print (len(words),"unique words:\n\n", words)
```

## Generating Continuous Bag of Words (CBOW)

```python
train = []
target = []

empty_output = [0 for _ in range(len(tags))]

for x, doc in enumerate(doc_words):
    bag = []
    
    w_temp = [stemmer.stem(w.lower()) for w in doc]
    
    for w in words:
        if w in w_temp:
            bag.append(1)
        else:
            bag.append(0)
            
    output_row = empty_output[:]
    output_row[tags.index(doc_tags[x])]=1
    
    train.append(bag)
    target.append(output_row)
    
```

```python
train = np.array(train)
```

```python
target = np.array(target)
```

## Building Neural Network Architecture

```python
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout

model = Sequential()
model.add(Dense(128,input_dim = len(train[0]), activation='relu'))
model.add(Dense(90, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(len(target[0]), activation='softmax'))
model.summary()
```

## Model Training

```python
np.array(train).shape, np.array(target).shape
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

history = model.fit(train, target, epochs=1000, verbose=1)
model.save('chatbot_model.h5', history)
```

## Loading the intents corpus, neural network model, words and tags

```python
from keras.models import load_model
import random
import pickle
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

model = load_model('chatbot_model.h5')
intents = json.loads(open(r'GL+Bot.json').read())
words = pickle.load(open('words.pkl','rb'))
tags = pickle.load(open('tags.pkl','rb'))
```

```python
def clean_up_sentence(sentence):
    sentence_words = nltk.word_tokenize(sentence)
    sentence_words = [lemmatizer.lemmatize(word.lower()) for word in sentence_words]
    return sentence_words

def bow(sentence, words, show_details=True):
    sentence_words = clean_up_sentence(sentence)
    bag = [0]*len(words) 
    for s in sentence_words:
        for i,w in enumerate(words):
            if w == s: 
                bag[i] = 1
                if show_details:
                    print ("found in bag: %s" % w)
    return(np.array(bag))

def predict_class(sentence, model):
    p = bow(sentence, words,show_details=False)
    res = model.predict(np.array([p]))[0]
    ERROR_THRESHOLD = 0.25
    results = [[i,r] for i,r in enumerate(res) if r>ERROR_THRESHOLD]
    # sort by strength of probability
    results.sort(key=lambda x: x[1], reverse=True)
    return_list = []
    for r in results:
        return_list.append({"intent": tags[r[0]], "probability": str(r[1])})
    return return_list
```

```python
def getResponse(ints, intents_json):
    tag = ints[0]['intent']
    list_of_intents = intents_json['intents']
    for i in list_of_intents:
        if(i['tag']== tag):
            result = random.choice(i['responses'])
            break
    return result
def chatbot_response(text):
    ints = predict_class(text, model)
    res = getResponse(ints, intents)
    return res
```

```python
def chat():
    print("Hello, Hope you're doing well. is there anything you're looking for?'")
    while(True):
        query = input("\n\nYou: ").lower()

        if query not in ['stop','exit']:
            print("Chatbot: ", end="")
            print(chatbot_response(query))
        else:
            print("See You Again")
            break
```
